```python
import pandas as pd
from shutil import copy
import os

df = pd.read_csv(r"E:\annotation_train_joints.csv")

categoryEmotions = ["Peace", "Affection", "Esteem", "Anticipation", "Engagement", "Confidence", 
                    "Happiness", "Pleasure", "Excitement", "Surprise", "Sympathy", "Doubt or Confusion", 
                    "Disconnect", "Fatigue", "Embarrassment", "Yearning", "Disapproval", "Aversion", 
                    "Annoyance", "Anger", "Sensitivity", "Sadness", "Disquietment", "Fear", "Pain", 
                    "Suffering"]

emotion_paths = r"E:\emotion_joints"

# Iterate over each row 
for index, row in df.iterrows():
    video_path = row["Video Path"]
    emotions = row["Emotion Value"].split(", ")  # Split emotion values
    
    # Iterate through each emotion row
    for emotion in emotions:
        emotion_folder = os.path.join(emotion_paths, emotion)
        
        # Create emotion folder if it doesn't exist
        if not os.path.exists(emotion_folder):
            os.makedirs(emotion_folder)

        video_file = os.path.basename(video_path)
        
        # Construct destination path
        destination_path = os.path.join(emotion_folder, video_file)
        
        # Copy the file directly to the emotion folder
        try:
            copy(video_path, destination_path)
            print(f"Copied {video_path} to {destination_path}")
        except Exception as e:
            print(f"Error copying {video_path} to {destination_path}: {e}")
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv(r"E:\annotation_train_joints.csv")

# Replace "mp4" with "npy"
def replace_last_part(value):
    parts = value.split('/')
    last_part = parts[-1]
    if last_part.endswith('.mp4'):
        new_last_part = last_part.replace('.mp4', '.npy')
        parts[-1] = new_last_part
        return '/'.join(parts)
    else:
        return value

df['Video Path'] = df['Video Path'].apply(replace_last_part)

df.to_csv(r'E:\updated_annotation_train_joints.csv', index=False)

print('CSV file updated successfully.')
```
